```python
# returns time series for an unskewed gaussian
# with time series x, amplitude a, center index b, and width c.
def gaussian(x, a, b, c):
    return a*np.exp(c*((x-b)**2))

# returns time series for a skewed gaussian
# with time series x, amplitude a, center index b, and 'negative narrowness' c,
# and skew parameter d
def skewed_gaussian(x, a, b, c, d):
    center = b
    skew_param = d
    normal_gaussian = gaussian(x, a, b, c)
    cdf = norm.cdf(skew_param*(x-center))
    skewed_function = 2*cdf*normal_gaussian
    return skewed_function
```

```python
signals.keys()
```

```python
params, _ = curve_fit(f=skewed_gaussian,
                    xdata=np.linspace(0, len(ecg_cycle), len(ecg_cycle)),
                    ydata=ecg_cycle,
                    # guesses
                    # amp(a)=2000, center(b)=peak guess, width(c)=100, skew(d)=0
                    p0=[12000, info['ECG_R_Peaks'][1], -.009, 0], 
                    maxfev=10000,
                    # lower bounds:
                    # amp(a)=-np.inf, center(b)=600, width(c)=-np.inf, skew(d)=-np.inf
                    # upper bounds:
                    # amp(a)=np.inf, center(b)=800, width(c)=np.inf, skew(d)=np.inf
                    bounds=([-np.inf,1700,-np.inf,-np.inf],[np.inf,1800,np.inf,np.inf]))
```

```python
params
```

```python
fit_func = skewed_gaussian(time_cycle, *params)
```

```python
plt.plot(time_cycle, ecg_cycle)
plt.plot(time_cycle, fit_func, linestyle='--')
plt.show()
```
